```python
#Task 2 - qn 2
import matplotlib.pyplot as plt

# Generate distributions
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plotting
plt.figure(figsize=(10, 6))

plt.hist(uniform_data, density=True, bins=80, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=80, histtype="step", color="red", label="Normal")
plt.hist(exp_data, density=True, bins=80, histtype="step", color="green", label="Exponential")

plt.xlim([-2, 8])  # Adjusted to visualize all three distributions clearly
plt.ylim([0, 1.2])
plt.legend(loc="upper right")
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#Task 1 - qn 3
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define range and meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Set limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([0, 50])

plt.show()
```

```python
#Task 1 - qn 4

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Task 4: Correlation analysis for Pokemon data
print("Task 4: Correlation analysis for Pokemon stats")
print("=" * 50)

# Load the Pokemon dataset (commonly used in seaborn tutorials)
try:
    # Try to load from seaborn's built-in datasets first
    pokemon_url = "https://raw.githubusercontent.com/datasets/pokemon/master/data/pokemon.csv"
    pokemon_df = pd.read_csv(pokemon_url)

    # Select the required features
    features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

    # Check if columns exist and rename if necessary
    if 'Sp. Atk' not in pokemon_df.columns and 'Special Attack' in pokemon_df.columns:
        pokemon_df = pokemon_df.rename(columns={'Special Attack': 'Sp. Atk', 'Special Defense': 'Sp. Def'})

    # Extract the features for correlation analysis
    pokemon_stats = pokemon_df[features].dropna()

    print(f"Dataset shape: {pokemon_stats.shape}")
    print(f"Features analyzed: {features}")
    print()

    # Calculate Pearson correlation coefficient
    pearson_corr = pokemon_stats.corr(method='pearson')

    # Calculate Spearman rank correlation
    spearman_corr = pokemon_stats.corr(method='spearman')

    # Create visualizations
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Pearson correlation heatmap
    sns.heatmap(pearson_corr, annot=True, cmap='inferno', center=0,
                square=True, fmt='.3f', cbar_kws={"shrink": .8}, ax=axes[0])
    axes[0].set_title('Pearson Correlation Coefficient\n(Linear Relationships)',
                      fontsize=12, fontweight='bold')

    # Spearman correlation heatmap
    sns.heatmap(spearman_corr, annot=True, cmap='inferno', center=0,
                square=True, fmt='.3f', cbar_kws={"shrink": .8}, ax=axes[1])
    axes[1].set_title('Spearman Rank Correlation\n(Monotonic Relationships)',
                      fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Print correlation matrices
    print("Pearson Correlation Matrix:")
    print(pearson_corr.round(3))
    print("\nSpearman Rank Correlation Matrix:")
    print(spearman_corr.round(3))

    # Highlight strongest correlations
    print("\nStrongest Pearson correlations (excluding diagonal):")
    pearson_flat = pearson_corr.where(np.triu(np.ones(pearson_corr.shape), k=1).astype(bool))
    pearson_pairs = pearson_flat.stack().sort_values(ascending=False)
    for pair, corr in pearson_pairs.head(3).items():
        print(f"{pair[0]} - {pair[1]}: {corr:.3f}")

    print("\nStrongest Spearman correlations (excluding diagonal):")
    spearman_flat = spearman_corr.where(np.triu(np.ones(spearman_corr.shape), k=1).astype(bool))
    spearman_pairs = spearman_flat.stack().sort_values(ascending=False)
    for pair, corr in spearman_pairs.head(3).items():
        print(f"{pair[0]} - {pair[1]}: {corr:.3f}")

except Exception as e:
    print(f"Could not load Pokemon dataset: {e}")
    print("Creating synthetic data for demonstration...")

    # Create synthetic Pokemon-like data
    np.random.seed(42)
    n_pokemon = 800

    # Generate correlated features that mimic Pokemon stats
    base_stats = np.random.exponential(50, n_pokemon)

    pokemon_synthetic = pd.DataFrame({
        'HP': base_stats + np.random.normal(0, 15, n_pokemon),
        'Attack': base_stats * 1.2 + np.random.normal(0, 20, n_pokemon),
        'Defense': base_stats * 0.9 + np.random.normal(0, 18, n_pokemon),
        'Sp. Atk': base_stats * 1.1 + np.random.normal(0, 22, n_pokemon),
        'Sp. Def': base_stats * 0.8 + np.random.normal(0, 16, n_pokemon),
        'Speed': np.random.exponential(60, n_pokemon) + np.random.normal(0, 25, n_pokemon)
    })

    # Ensure positive values
    pokemon_synthetic = pokemon_synthetic.clip(lower=1)

    # Calculate correlations
    pearson_corr = pokemon_synthetic.corr(method='pearson')
    spearman_corr = pokemon_synthetic.corr(method='spearman')

    # Create visualizations
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    sns.heatmap(pearson_corr, annot=True, cmap='inferno', center=0,
                square=True, fmt='.3f', cbar_kws={"shrink": .8}, ax=axes[0])
    axes[0].set_title('Pearson Correlation Coefficient\n(Synthetic Pokemon Data)',
                      fontsize=12, fontweight='bold')

    sns.heatmap(spearman_corr, annot=True, cmap='inferno', center=0,
                square=True, fmt='.3f', cbar_kws={"shrink": .8}, ax=axes[1])
    axes[1].set_title('Spearman Rank Correlation\n(Synthetic Pokemon Data)',
                      fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.show()

    print("Pearson Correlation Matrix (Synthetic Data):")
    print(pearson_corr.round(3))
    print("\nSpearman Rank Correlation Matrix (Synthetic Data):")
    print(spearman_corr.round(3))

print("\nAll tasks completed successfully!")
```

```python
#Task 2 - qn 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')

# Create a new dataframe with specified columns
filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
display(filtered_df.head(21))
```

```python
# display(filtered_df.describe())
summary_stats = filtered_df.describe()
mean_values = filtered_df.mean()
std_values = filtered_df.std()

print("Mean:\n", mean_values)
print("\nStandard Deviation:\n", std_values)
print("\nQuartiles:\n", summary_stats.loc[['25%', '50%', '75%']])
```

```python
plt.figure(figsize=(20, 4))
for idx, column in enumerate(filtered_df.columns):
    plt.subplot(1, 4, idx + 1)
    sns.histplot(filtered_df[column], kde=True)
    plt.title(f'{column} Histogram')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(filtered_df)
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```
